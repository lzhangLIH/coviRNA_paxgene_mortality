```python
# apply threshold to positive probabilities to create labels
def to_labels(pos_probs, threshold):
 return (pos_probs >= threshold).astype('int')
```

```python
scoring = 'f1'
filename = 'metrics_quebec'

eval_metrics = pd.DataFrame(columns=["classifier","auroc", "auprc", "accuracy",
                                         "balanced accuracy", "mcc",
                                         "precision","precision_0", "precision_1", 
                                         "recall","recall_0", "recall_1", 
                                         "f1","f1_0", "f1_1",'brier'])

for key, clf in lst_clf.items():
    print(key)
    
    # hyperparameter tuning
    pipeline = Pipeline([('scale', PowerTransformer()), #####
                            # ("re_s", RandomUnderSampler(random_state=random_state)),  #####
                         # ('clf', CalibratedClassifierCV(clf, cv=5))])
                          ("clf", clf)])

    grid = lst_grid[key]

    # search = RandomizedSearchCV(
    #     pipeline, grid, scoring=metrics.make_scorer(metrics.matthews_corrcoef), n_iter=100,random_state=random_state, ###############
    #     n_jobs=n_jobs, cv=kf,return_train_score=True
    # ).fit(X_train, y_train)

    search = GridSearchCV(
        pipeline, grid, 
        # scoring=metrics.make_scorer(metrics.matthews_corrcoef),
        scoring=scoring, 
        n_jobs=n_jobs, cv=kf,return_train_score=True
    ).fit(X_train_, y_train_)

    best_score = search.best_score_
    print(f"Best Tuning {scoring} score: {best_score}") ###############

    best_params = {
        key: value for key, value in search.best_params_.items()
    }
    print(best_params)

    best_estimator = search.best_estimator_ # a refitted pipeline here
    
    # # get the best threshold
    # probs_train = best_estimator.predict_proba(X_train_)[:,1]
    # fpr_train, tpr_train, thresholds_train = metrics.roc_curve(y_train_, probs_train)
    # J = tpr_train - fpr_train
    # ix = np.argmax(J)
    # best_thresh = thresholds_train[ix]
    # print('Best Threshold=%f' % (best_thresh))
    
    y_probs = best_estimator.predict_proba(X_test_)[:,1]
    
    fp_rate, tp_rate, thresholds = metrics.roc_curve(y_test_, y_probs)
    auc = metrics.auc(fp_rate, tp_rate)
    
    # area under precision-recall curve
    precision_, recall_, thresholds_ = metrics.precision_recall_curve(y_test_, y_probs)
    auc_pr = metrics.auc(recall_, precision_)
    
    # convert to f score
    fscore = (2 * precision_ * recall_) / (precision_ + recall_)
    # locate the index of the largest f score
    ix = argmax(fscore)
    best_thresh = thresholds_[ix]
    print('Best Threshold=%f' % (best_thresh))
    
    # y_pred = best_estimator.predict(X_test_)
    y_pred = to_labels(y_probs, best_thresh)

    ave = 'macro'

    precision = metrics.precision_score(y_test_, y_pred, average=ave)
    precision_1 = metrics.precision_score(y_test_, y_pred)
    precision_0 = metrics.precision_score(y_test_, y_pred, pos_label=0)
    f1 = metrics.f1_score(y_test_, y_pred, average=ave)
    f1_1 = metrics.f1_score(y_test_, y_pred)
    f1_0 = metrics.f1_score(y_test_, y_pred, pos_label=0)
    accuracy = metrics.accuracy_score(y_test_, y_pred)
    mcc = metrics.matthews_corrcoef(y_test_, y_pred)
    recall = metrics.recall_score(y_test_, y_pred, average=ave)
    recall_1 = metrics.recall_score(y_test_, y_pred)
    recall_0 = metrics.recall_score(y_test_, y_pred, pos_label=0)
    balanced_accuracy = metrics.balanced_accuracy_score(y_test_, y_pred)

    brier = metrics.brier_score_loss(y_test_, y_probs)

    eval_metrics = pd.concat([eval_metrics,pd.DataFrame.from_dict({'classifier':key,
                                                                   'auroc': auc,
                                                                   'auprc': auc_pr,
                                                                    'precision': precision,
                                                                    'precision_1': precision_1,
                                                                    'precision_0': precision_0,
                                                                    'f1': f1,
                                                                    'f1_1': f1_1,
                                                                    'f1_0': f1_0,
                                                                    'accuracy': accuracy,
                                                                    'mcc': mcc,
                                                                    'recall': recall,
                                                                    'recall_1': recall_1,
                                                                    'recall_0': recall_0,
                                                                    'balanced accuracy': balanced_accuracy,
                                                                    'brier': brier}, orient='index').transpose()], ignore_index=True)    
    file_out = filename+'.csv' ###############
    eval_metrics.to_csv(os.path.join(folder_out,file_out), index=False)
```
